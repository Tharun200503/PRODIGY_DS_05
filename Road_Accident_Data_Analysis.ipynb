```python
df = df.drop_duplicates()
```

```python
sns.countplot(x = df['Accident_Severity'])
plt.title('Distribution of Accident Severity')
```

```python
df.isna().sum()

df.drop(['Service_year_of_vehicle','Defect_of_vehicle','Work_of_casuality', 'Fitness_of_casuality','Time'],
        axis = 1, inplace = True)
df.head()
```

```python
df.drop(['Time','Carriageway_Hazards'],
        axis = 1, inplace = True)
df.head()
```

```python
categorical=[i for i in df.columns if df[i].dtype=='O']
print('The categorical variables are',categorical)
```

```python
for i in categorical:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df.isna().sum()
```

```python
sns.countplot(x='Accident_Severity', hue='Road_Surface_Conditions', data=df)
plt.title('Accident Severity by Road Surface Condition')
plt.show()
```

```python
sns.countplot(x='Accident_Severity', hue='Road_Type', data=df)
plt.title('Accident Severity by Road Type')
plt.show()
```

```python
sns.countplot(x='Accident_Severity', hue='Weather_Conditions', data=df)
plt.title('Accident Severity by Weather Conditions')
plt.show()
```

```python
sns.countplot(x='Accident_Severity', hue='Vehicle_Type', data=df)
plt.title('Accident Severity by Vehicle Type')
plt.show()
```

```python
sns.countplot(x='Accident_Severity', hue='Day_of_Week', data=df)
plt.title('Accident Severity based on Days')
plt.show()
```

```python
import geopandas as gpd
#identify hotspots
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))
hotspots = gdf.groupby(['Longitude', 'Latitude']).size().reset_index(name='count')
hotspots = hotspots.sort_values('count', ascending=False).head(20)
```
